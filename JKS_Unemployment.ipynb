```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
import scipy.stats as st
```

```python
unemployment_csv = 'clean_data/unemployment_df.csv'
unemployment_df = pd.read_csv(unemployment_csv)
unemployment_df
```

```python
column_1 = unemployment_df["Female - 2018"]

column_2 = unemployment_df["Male - 2018"]

correlation = column_1.corr(column_2)

print(correlation)
```

```python
unemployment_data = unemployment_df.drop(labels='Total - 2018', axis=1)
unemployment_data
```

```python
unemployment_data.drop(unemployment_data.index[range(0)], axis=0, inplace=True)
unemployment_data
```

```python
#multi_plot.title('Unemployment gender differences per country(%)', color="chartreuse")

# Set x axis and tick locations
x_axis = unemployment_data["Country Name"]
tick_locations = [value for value in x_axis]

# The colors of each section of the bar chart
color = ["hotpink", "lawngreen"]

# Create a list indicating where to write x labels and set figure size to adjust for space
multi_plot = unemployment_data.plot(kind="bar", figsize=(10,8), color=color, title = "Unemployment gender differences per country(%)")
multi_plot.set_xticklabels(unemployment_data["Country Name"], rotation=45)

plt.grid()
plt.savefig('Unemployment_Gender_Differences')
```

```python
unemployment_data
```

```python
# using subplots() to draw vertical lines
fig, axes = plt.subplots(figsize=(10,7))

#fig, ax = plt.subplots(2, 2, figsize=(10,7))
#fig.tight_layout()

axes.vlines(unemployment_data.index, ymin=0, ymax=unemployment_data['Female - 2018'])
axes.vlines(unemployment_data.index, ymin=0, ymax=unemployment_data['Male - 2018'])
#axes.vlines(unemployment_data.index, figsize=(12,8)
# drawing the markers (circle)
axes.plot(unemployment_data["Country Name"], unemployment_data['Female - 2018'], "o")
axes.plot(unemployment_data["Country Name"], unemployment_data['Male - 2018'], "^")
axes.set_ylim(0)
line_colors = ['blue', 'cyan', 'green', 'red',
               'skyblue', 'brown', 'yellow',
               'black', 'grey', 'orange', 'maroon',
               'lightgreen']
axes.vlines(unemployment_data.index, ymin=0,
            ymax=unemployment_data['Female - 2018'], colors=line_colors)

plt.gca().legend(('Female - 2018', 'Male - 2018'))
# formatting and details
plt.xlabel('Country Name')
plt.ylabel('Percent Unemployment')
plt.title('Youth Age 15-24 Unemployment per country(%), 2018')
plt.xticks(unemployment_data["Country Name"])
plt.xticks(rotation=45)
plt.savefig('new Youth Age 15-24 Unemployment per country(%), 2018')
plt.show()
```
