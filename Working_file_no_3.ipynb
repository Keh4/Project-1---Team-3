```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
import os
import numpy
import requests as rq
import pandas as pd
from io import BytesIO
import json

```

```python
csv_path = os.path.join('raw_data/WDI_Trimmed_Dataset_20220130.csv')
with open(csv_path) as csvfile:
    trimmed_df = pd.read_csv(csv_path, header=[0])
    trimmed_df = pd.DataFrame(trimmed_df)
```

```python
trimmed_df
```

```python
print("**************************")
print("GDP in Current US Dolloars")
print("**************************")

gdp_df = trimmed_df.loc[trimmed_df['Series Name'] == 'GDP (current US$)']
gdp_df = gdp_df.drop(columns='Series Code')
gdp_df = gdp_df.drop(columns='Country Code')
gdp_df = gdp_df.drop(columns='Series Name')
gdp_df = gdp_df.set_index('Country Name')
gdp_df = gdp_df.drop(index='Country List')
gdp_df
```

```python
print("**************************")
print("2018 Human Capital Index")
print("**************************")

hci_df = trimmed_df.loc[trimmed_df['Series Name'] == 'Human capital index (HCI) (scale 0-1)']
hci_df = hci_df.drop(columns='Series Code')
hci_df = hci_df.drop(columns='Country Code')
hci_df = hci_df.drop(columns='Series Name')
hci_df = hci_df.drop(columns='2014 [YR2014]')
hci_df = hci_df.drop(columns='2015 [YR2015]')
hci_df = hci_df.drop(columns='2016 [YR2016]')
hci_df = hci_df.drop(columns='2017 [YR2017]')
hci_df = hci_df.drop(columns='2019 [YR2019]')
hci_df = hci_df.drop(columns='2020 [YR2020]')
hci_df = hci_df.drop(columns='2010-2020')
hci_df = hci_df.set_index('Country Name')
hci_df = hci_df.drop(index='Country List')
hci_df
```

```python
# Isolate female labor force participation
labor_f_df = trimmed_df.loc[trimmed_df['Series Code'] == 'SL.TLF.ACTI.1524.FE.ZS']
labor_f_df = labor_f_df.set_index('Country Name')
labor_f_df = labor_f_df.drop(columns='Series Code')
labor_f_df = labor_f_df.drop(columns='Country Code')
labor_f_df = labor_f_df.drop(columns='Series Name')
labor_f_df = labor_f_df.drop(columns='2020 [YR2020]')
labor_f_df = labor_f_df.drop(columns='2010-2020')
labor_f_df = labor_f_df.drop(index='Country List')

# Isolate male labor force participation
labor_m_df = trimmed_df.loc[trimmed_df['Series Code'] == 'SL.TLF.ACTI.1524.MA.ZS']
labor_m_df = labor_m_df.set_index('Country Name')
labor_m_df = labor_m_df.drop(columns='Series Code')
labor_m_df = labor_m_df.drop(columns='Country Code')
labor_m_df = labor_m_df.drop(columns='Series Name')
labor_m_df = labor_m_df.drop(columns='2020 [YR2020]')
labor_m_df = labor_m_df.drop(columns='2010-2020')
labor_m_df = labor_m_df.drop(index='Country List')
```

```python
# Isolate employers data
employers_f_df = trimmed_df.loc[trimmed_df['Series Code'] == 'SL.EMP.MPYR.FE.ZS']
employers_f_df = employers_f_df.drop(['Country Code',
                                      'Series Code',
                                      'Series Name',
                                      '2014 [YR2014]',
                                      '2015 [YR2015]',
                                      '2016 [YR2016]',
                                      '2017 [YR2017]',
                                      '2019 [YR2019]',
                                      '2020 [YR2020]',
                                      '2010-2020'],
                                     axis = 1)
employers_f_df = employers_f_df.rename(columns={"2018 [YR2018]": "Feamle - 2018"})
employers_f_df = employers_f_df.set_index("Country Name")
employers_f_df = employers_f_df.drop(index='Country List')




employers_m_df = trimmed_df.loc[trimmed_df['Series Code'] == 'SL.EMP.MPYR.MA.ZS']
employers_m_df = employers_m_df.drop(['Country Code',
                                      'Series Code',
                                      'Series Name',
                                      '2014 [YR2014]',
                                      '2015 [YR2015]',
                                      '2016 [YR2016]',
                                      '2017 [YR2017]',
                                      '2019 [YR2019]',
                                      '2020 [YR2020]',
                                      '2010-2020'],
                                     axis = 1)
employers_m_df = employers_m_df.rename(columns={"2018 [YR2018]": "Male - 2018"})
employers_m_df = employers_m_df.set_index("Country Name")
employers_m_df = employers_m_df.drop(index='Country List')






employers_total_df = trimmed_df.loc[trimmed_df['Series Code'] == 'SL.EMP.MPYR.ZS']
employers_total_df = employers_total_df.drop(['Country Code',
                                      'Series Code',
                                      'Series Name',
                                      '2014 [YR2014]',
                                      '2015 [YR2015]',
                                      '2016 [YR2016]',
                                      '2017 [YR2017]',
                                      '2019 [YR2019]',
                                      '2020 [YR2020]',
                                      '2010-2020'],
                                     axis = 1)
employers_total_df = employers_total_df.rename(columns={"2018 [YR2018]": "Total - 2018"})
employers_total_df = employers_total_df.set_index("Country Name")
employers_total_df = employers_total_df.drop(index='Country List')
```
