```python
#remove unneccesary column <Total - 2018>
population_total_data = population_df.drop(labels='Female - 2018', axis=1)
population_total = population_total_data.drop(labels='Male - 2018', axis=1)
population_total
```

```python
#multi_plot.title('Unemployment gender differences per country(%)', color="chartreuse")

# Set x axis and tick locations
x_axis = population_total["Country Name"]
tick_locations = [value for value in x_axis]

# Create a list indicating where to write x labels and set figure size to adjust for space
multi_plot = population_total.plot(kind="bar", figsize=(10,10), color="gold", title = "Total population per country, year 2018")
multi_plot.set_xticklabels(population_total["Country Name"], rotation=45)
  
# Setting the background color of the plot 
# using set_facecolor() method

plt.grid()
```
